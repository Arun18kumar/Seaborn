```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data = tips)
```

```python
sns.barplot(x='sex',y='total_bill',data = tips,estimator=np.std)

# estimator estimates the function passed as the argument. We can also pass user defined funtion as a argument
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='total_bill',y='day',data=tips)
```

```python
sns.boxplot(x='total_bill',y='day',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)

# Violin plot is same as that of the box plot but, it will give more specific graphical diagram based on the data 
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)

# If the graph has a single dotted output, then add an argument (jitter = True).
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)

# swarmplot is a combination of both strip and violin plots (Not to check swarmplot for large datasets )
```

```python
#  Combination of violin plot and swarm plot

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')

#  This plot is mostly used for data analysis not for presentation
```
