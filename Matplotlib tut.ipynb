```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#  Functional

plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,2)
plt.plot(x,y,'r')

plt.subplot(1,2,1)
plt.plot(y,x,'b')
```

```python
# Object Oriented

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
```
