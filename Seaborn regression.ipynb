```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')

# lmplot accesses matplotlib functions most of the arguments used in this plot is matplolib's functionalities
# lmplot is also a regression plot
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])

#  makers will differentiate the data
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v']
          , scatter_kws={'s':70})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')

#  the argument col makes a categorical grided plot base upon the categories
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')

# graphs are too dense to interpret to understand we can add aspect_ratio
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex'
          ,aspect=0.5,size=8)
```
